# Data Preprocessing

```python
data
```

```python
dumms1 = pd.get_dummies(data['Real time GPS Data'])
data = pd.concat([data, dumms1], axis =1)
data.drop(['Real time GPS Data', 'Not Congested'], inplace=True, axis=1)
```

```python
dumms1 = pd.get_dummies(data['Real time weather data'])
data = pd.concat([data, dumms1], axis =1)
data.drop(['Real time weather data', 'Cloudy'], inplace=True, axis=1)
```

```python
data.drop(['Delivery Location'], inplace = True, axis =1)
```

```python
mapping = {'N': 0, 'Y':1}
```

```python
data['Delivery Delay?'] = data['Delivery Delay?'].map(mapping)
```

```python
dumms1 = pd.get_dummies(data['Type of Product'])
data = pd.concat([data, dumms1], axis =1)
data.drop(['Type of Product', 'Condition Products (Perishable, Requiring Ventilation, Etc.)'], inplace=True, axis=1)
```

```python
dumms1 = pd.get_dummies(data['Type of terrain of delivery location'])
data = pd.concat([data, dumms1], axis =1)
data.drop(['Type of terrain of delivery location', 'Flat To Undulating Plains'], inplace=True, axis=1)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn import metrics
```

```python
data['Delivery Delay?']
```

```python
X = data.drop(['Delivery Delay?'], axis =1)
```

```python
y = data['Delivery Delay?']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 5)
```

```python
X_train.shape, y_train.shape
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
print(cross_val_score(model, X, y, cv = 10, scoring = 'accuracy').mean())
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
print(cross_val_score(rfc, X, y, cv=10, scoring = 'accuracy').mean())
```
